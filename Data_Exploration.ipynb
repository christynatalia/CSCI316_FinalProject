```python
from pyspark.sql import SparkSession
spark = SparkSession \
    .builder \
    .config("spark.csci316.data.exploration", "false") \
    .getOrCreate()
spark
```

```python
#Read the csv file
data = spark.read.csv("cleanWithHeader.csv", inferSchema=True, header=True)
```

```python
pd_data = data.toPandas()
```

```python
pd_data.hist('default_ind')
```
